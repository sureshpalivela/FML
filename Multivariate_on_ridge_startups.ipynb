```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import linear_model
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
```

```python
ds=pd.read_csv("/content/drive/MyDrive/50_Startups.csv")
ds
```

```python
ds.corr()
```

```python
x_profit=ds.drop('Profit', axis=1)
y = ds['Profit'].values
```

```python
x_profit=ds.drop('State',axis=1)
```

```python
x_profit
```

```python
x_profit_train, x_profit_test, y_train, y_test = train_test_split(x_profit, y, test_size=0.25, random_state=42)
```

```python
x_profit_train.shape, x_profit_test.shape
```

```python
reg = Ridge(alpha = 0.1)
reg.fit(x_profit_train, y_train)
```

training a model using training data 

```python
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
y_pred1 = reg.predict(x_profit_train)
print(f'MAE = {mean_absolute_error(y_train, y_pred1)}')
print(f'MSE = {mean_squared_error(y_train, y_pred1)}')
print(f'RMSE = {np.sqrt(mean_squared_error(y_train, y_pred1))}')
print(f'R_2 = {r2_score(y_train, y_pred1)}')
```

traning a model using testing data

```python
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
y_pred2 = reg.predict(x_profit_test)
print(f'MAE = {mean_absolute_error(y_test, y_pred2)}')
print(f'MSE = {mean_squared_error(y_test, y_pred2)}')
print(f'RMSE = {np.sqrt(mean_squared_error(y_test, y_pred2))}')
print(f'R_2 = {r2_score(y_test, y_pred2)}')
```
