```python
cm=confusion_matrix(X,y)
print(cm)
```

```python
ax=plt.axes()
ds_cm=cm
sns.heatmap(ds_cm,annot=True,annot_kws={"size":30},fmt='d',cmap="Blues",ax=ax)
ax.set_title("confusion matrix")
plt.show()
```
